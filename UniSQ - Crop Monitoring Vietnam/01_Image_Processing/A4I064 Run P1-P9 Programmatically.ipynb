```python
import os
import papermill as pm

crop_names = ["Rice", "Mango", "Dragonfruit"]
```

```python
def RunScripts1234(script, source_folder):
    P_dates = [f.name for f in os.scandir(source_folder) if f.is_dir()]
    print(P_dates)
    
    for date in P_dates:
        pm.execute_notebook(script, f'Papermill/{os.path.basename(script)}_{date}.ipynb', 
                            parameters=dict(source_folder=source_folder, date_str=date))
```

```python
def RunScripts5689(script, source_folder):
    for crop in crop_names:
        pm.execute_notebook(script, f'Papermill/{os.path.basename(script)}_{crop}.ipynb', 
                            parameters=dict(source_folder=source_folder, crop_name=crop))
```

```python
def RunScripts7(script, source_folder):
    parent_folders = [f.name for f in os.scandir(source_folder) if f.is_dir()]
    print(parent_folders)
    
    for folder in parent_folders:
        cycles = [f.name for f in os.scandir(os.path.join(source_folder,folder)) if f.is_dir()]
        
        for cycle in cycles:        
            pm.execute_notebook(script, f'Papermill/{os.path.basename(script)}_{folder}_{cycle}.ipynb', 
                                parameters=dict(source_folder=source_folder, parent_folder=folder, cycle=cycle))
```
